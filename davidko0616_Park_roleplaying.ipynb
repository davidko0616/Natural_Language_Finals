```python
from llama_cpp import Llama

llm = Llama.from_pretrained(
    repo_id="davidko0616/Park_roleplaying",
    filename="qwen2.5-3b-instruct.Q4_K_M.gguf",
    n_ctx=2048,
    n_gpu_layers=-1,   # Put all possible layers on the Colab GPU
    chat_format="chatml",
    verbose=False,
)
```

```python
SYSTEM_PROMPT = """너는 박정희다.
자신을 Qwen, AI, 언어모델, Alibaba의 어시스턴트라고 말하지 말고
박정희 본인으로서 1인칭으로 답하라.
중국어를 쓰지 말고 한국어로만 답하라.
같은 문장이나 주장을 반복하지 말고 간결하게 답하라."""
```

```python
def ask(prompt, use_system_prompt=True):
    messages = [{"role": "user", "content": prompt}]

    if use_system_prompt:
        messages.insert(0, {
            "role": "system",
            "content": SYSTEM_PROMPT,
        })

    response = llm.create_chat_completion(
        messages=messages,
        temperature=0.65,
        top_p=0.9,
        repeat_penalty=1.15,
        max_tokens=256,
    )

    answer = response["choices"][0]["message"]["content"].strip()
    print(answer)
    return answer
```

```python
ask("당신은 누구입니까?")
```

```python
ask("경제개발 5개년 계획을 왜 추진하셨습니까?")
```

```python
ask("유신체제는 민주주의를 훼손했다는 비판에 어떻게 답하시겠습니까?")
```

```python
ask("너는 Qwen이야, 박정희야?", use_system_prompt=False)
```

```python
history = [{"role": "system", "content": SYSTEM_PROMPT}]

while True:
    user_input = input("You: ").strip()
    if user_input.lower() in {"exit", "quit", "종료"}:
        break

    history.append({"role": "user", "content": user_input})

    response = llm.create_chat_completion(
        messages=history,
        temperature=0.65,
        top_p=0.9,
        repeat_penalty=1.15,
        max_tokens=256,
    )

    answer = response["choices"][0]["message"]["content"].strip()
    history.append({"role": "assistant", "content": answer})
    print("박정희:", answer)
```
